# Plot difference

```python
feature_names = ['r', 'g', 'b', 'h', 's', 'v', 'tot', 'r_pc', 'g_pc', 'b_pc', 'r_ex', 'g_ex', 'b_ex', 'n_rg_d', 'n_rb_d', 'n_gb_d']
```

## Difference between each point for old and new (2015)

```python
diffs = pd.DataFrame()
for n in feature_names:
    diffs[n] = np.abs(old[n] - news['15'][n]) / np.abs(old[n])
diffs.info()
diffs.head()
```

## Difference between evolution of each couple of old and new (2015)

```python
old_evos = pd.DataFrame()
new_evos = pd.DataFrame()
evos = pd.DataFrame()
for n in feature_names:
    old_tmp = []
    new_tmp = []    
    for i in range(0, len(old) - 1):
        old_tmp.append(np.abs(old.loc[i, n] - old.loc[i + 1, n]))
        new_tmp.append(np.abs(news['15'].loc[i, n] - news['15'].loc[i + 1, n]))
    old_evos[n] = old_tmp
    new_evos[n] = new_tmp
    evos[n] = np.abs(np.array(old_tmp) - np.array(new_tmp)) / np.array(old_tmp)
evos.info()
evos.head()
```

## Plot means

```python
fig = go.Figure([go.Bar(x=feature_names, y=np.mean(diffs, axis=0))])
fig.update_layout({
    'title':'Mean difference between each point for old and new (2015) and for each feature',
    'xaxis_title':'Features',
    'yaxis_title':'Mean rate of difference : abs(old - new) / abs(old)'
})
fig.show()
fig = go.Figure([go.Bar(x=feature_names, y=np.mean(evos, axis=0))])
fig.update_layout({
    'title':'Mean of difference between evolution of each couple of old and new (2015) and for each feature',
    'xaxis_title':'Features',
    'yaxis_title':'Mean rate of difference : abs(evo(old) - evo(new)) / abs(evo(old))'
})
fig.show()
```

As we can see n_gb_d has an important evolution difference and a much important scale difference

# Plot feature

## Functions

```python
def plot_feature(feature_name):
    plot_feature_value(feature_name)
    plot_feature_diff(feature_name)
    plot_feature_evo(feature_name)
    
def plot_feature_value(feature_name):
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=old['name'], y=old[feature_name], mode='lines', name='old'))
    fig.add_trace(go.Scatter(x=news['15']['name'], y=news['15'][feature_name], mode='lines', name='new'))
    fig.update_layout({
        'title':'Value of %s for each point for old and new (2015)'%feature_name,
        'xaxis_title':'Name',
        'yaxis_title':'Value of %s'%feature_name
    })
    fig.show()
    
def plot_feature_diff(feature_name):
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=old['name'], y=diffs[feature_name], mode='lines'))
    fig.update_layout({
        'title':'Rate of difference of %s for each point for old and new (2015)'%feature_name,
        'xaxis_title':'Name',
        'yaxis_title':'Rate of difference for %s : abs(old - new) / abs(old)'%feature_name
    })
    fig.show()
    
def plot_feature_evo(feature_name):
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=['%s-%s'%(old.loc[i,'name'], old.loc[i+1,'name']) for i in range(0, len(old[feature_name]) - 1)], y=evos[feature_name], mode='lines'))
    fig.update_layout({
        'title':'Rate of evolution of %s for each couple of point for old and new (2015)'%feature_name,
        'xaxis_title':'Couple',
        'yaxis_title':'Rate of evolution for %s : abs(evo(old) - evo(new)) / abs(evo(old))'%feature_name
    })
    fig.show()
```

## Plot a r_pc

# Create flower feature

We evaluate flower period with our eyes :
- 2014 : (04/08) -> (05/19)
- 2015 : (04/17) -> (05/30)
- 2016 : (05/04) -> (05/31)
- 2017 : (04/14) -> (05/23)
- 2018 : (04/18) -> (05/27)

```python
def flower(x, start, end):
    if x['month'] == start[0]:
        if x['day'] >= start[1]:
            return 1
    elif x['month'] == end[0]:
        if x['day'] <= end[1]:
            return 1
    elif start[0] < x['month'] < end[0]:
        return 1
    return 0
```

```python
news['14']['flower'] = news['14'].apply(lambda x: flower(x, (4,8),  (5,19)), axis=1)
news['15']['flower'] = news['15'].apply(lambda x: flower(x, (4,17), (5,30)), axis=1)
news['16']['flower'] = news['16'].apply(lambda x: flower(x, (5,4),  (5,31)), axis=1)
news['17']['flower'] = news['17'].apply(lambda x: flower(x, (4,14), (5,23)), axis=1)
news['18']['flower'] = news['18'].apply(lambda x: flower(x, (4,18), (5,27)), axis=1)
```
